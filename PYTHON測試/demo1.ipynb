- 這裡是一個建立正弦波圖表的範例

```python
import numpy as np
import matplotlib.pyplot as plt

# 設定 x 軸的範圍，從 0 到 2π，間隔為 0.1
x = np.arange(0, 2 * np.pi, 0.1)

# 計算 y 值，即 sin(x) 的值
y = np.sin(x)

# 使用 plot() 函數繪製圖表
plt.plot(x, y)

# 設定圖表標題和軸標籤
plt.title('sin chart')
plt.xlabel('x')
plt.ylabel('sin(x)')

# 顯示圖表
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 設定 x 軸的範圍，從 -π/2 到 π/2，間隔為 0.1
x = np.arange(-np.pi/2 + 0.1, np.pi/2 - 0.1, 0.1)

# 計算 y 值，即 tan(x) 的值
y = np.tan(x)

# 使用 plot() 函數繪製圖表
plt.plot(x, y)

# 設定圖表標題和軸標籤
plt.title('tan chart')
plt.xlabel('x')
plt.ylabel('tan(x)')

# 設定 y 軸的範圍，避免圖表過於擁擠
plt.ylim(-10, 10)

# 顯示圖表
plt.show()
```
