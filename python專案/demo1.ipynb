- 這裡是一個建立正弦波圖表的範例

```python
import numpy as np
import matplotlib.pyplot as plt

# 設定 x 軸的範圍，從 0 到 2π，間隔為 0.1
x = np.arange(0, 2 * np.pi, 0.1)

# 計算 y 值，即 sin(x) 的值
y = np.sin(x)

# 使用 plot() 函數繪製圖表
plt.plot(x, y)

# 設定圖表標題和軸標籤
plt.title('sin chart')
plt.xlabel('x')
plt.ylabel('sin(x)')

# 顯示圖表
plt.show()
```

- 同時顯示sin, cos的正弦波和餘弦波的圖表

```python
import numpy as np
import matplotlib.pyplot as plt

# 設定 x 軸的範圍，從 0 到 2π，間隔為 0.1
x = np.arange(0, 2 * np.pi, 0.1)

# 計算 y 值，即 sin(x) 和 cos(x) 的值
y1 = np.sin(x)
y2 = np.cos(x)

# 使用 plot() 函數繪製圖表
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')

# 設定圖表標題和軸標籤
plt.title('sin and cos chart')
plt.xlabel('x')
plt.ylabel('value')

# 顯示圖例
plt.legend()

# 顯示圖表
plt.show()
```

- 餘弦波圖表

```python
import numpy as np
import matplotlib.pyplot as plt

# 設定 x 軸的範圍，從 0 到 2π，間隔為 0.1
x = np.arange(0, 2 * np.pi, 0.1)

# 計算 y 值，即 cos(x) 的值
y = np.cos(x)

# 使用 plot() 函數繪製圖表
plt.plot(x, y)

# 設定圖表標題和軸標籤
plt.title('cos plot')
plt.xlabel('x')
plt.ylabel('cos(x)')

# 顯示圖表
plt.show()
plt.show()
```
